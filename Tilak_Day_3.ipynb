## 1. What do you understand by Pruning in a Decision Tree and Why do we require Pruning in Decision Trees? Explain.

**Decision Tree Pruning:**

**Definition:** Decision tree pruning is a technique used to prevent decision trees from overfitting the training data.

**Objective:** Pruning simplifies the decision tree by removing non-critical and redundant parts, improving its ability to generalize to new data.

**Why Prune?:** Decision trees tend to become overly complex during training, capturing noise and memorizing the training data. Pruning helps create simpler, more robust models.

**Benefits of Decision Tree Pruning:**

Improves model interpretability.

Reduces overfitting.

Enhances generalization to unseen data.

## 2. How do you choose a machine learning algorithm for a problem statement? Explain Briefly.

Understand Your Problem:

Start by understanding your problem domain, data, and objectives.
Consider whether it’s a classification, regression, clustering, or other task.

Preprocess Data:

Clean, transform, and normalize your data.
Handle missing values, outliers, and categorical features.

Explore the Dataset:

Visualize data distributions, correlations, and patterns.
Identify relevant features and relationships.

Select Evaluation Metrics:

Choose appropriate metrics (accuracy, precision, recall, etc.) based on your problem.

Types of Algorithms:

Supervised Learning:

For labeled data (classification, regression).
Examples: Decision trees, logistic regression, SVM, neural networks.

Unsupervised Learning:

For unlabeled data (clustering, dimensionality reduction).
Examples: K-means, PCA, DBSCAN.

Semi-Supervised Learning:

Combines labeled and unlabeled data.
Examples: Self-training, co-training.

Reinforcement Learning:

Learn from interactions with an environment.
Examples: Q-learning, policy gradients.


## 3. What is Model Explainability and why is it required?


Model Explainability refers to the ability to understand and interpret how a machine learning model arrives at its predictions. It’s essential for several reasons:

Interpretability:

Human Trust: Users, stakeholders, and regulators need to trust AI systems. Explainable models are more transparent and easier to trust.
Legal and Ethical Compliance: Regulations (e.g., GDPR) often require explanations for automated decisions.

Debugging and Improvement:

Error Analysis: Understanding why a model fails helps improve it.
Feature Importance: Identifying influential features guides feature engineering.
Bias Detection: Explainability helps uncover biases in predictions.

Model Selection:

Comparing Models: Explainable models allow informed comparisons.
Trade-offs: Balance between accuracy and interpretability.

Domain-Specific Insights:

Medical Diagnosis: Doctors need to understand why a model diagnoses a disease.
Financial Risk Assessment: Explainability helps justify loan approvals or denials.

Algorithmic Fairness:

Detecting and mitigating bias requires understanding model decisions.
